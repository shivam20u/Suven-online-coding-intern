```python
df = weather[['Formatted Date','Apparent Temperature (C)','Humidity']]
df.head(10)
```

# Resample data from hourly to month wise

```python
df['Formatted Date']=pd.to_datetime(df['Formatted Date'], utc=True)
df=df.set_index('Formatted Date')
```

```python
df1 = df[['Apparent Temperature (C)','Humidity']].resample('MS').mean()
df1
```

# Analysis plots of Temperature & Humidity over the range of years in the dataset
Variation in Apparent Temperature & Humidity with time (in years)

```python
plt.figure(figsize=(16,9))
plt.plot(df1['Humidity'],label='Humidity',color='orange', linewidth = 3)
plt.plot(df1['Apparent Temperature (C)'],label='Apparent Temperature (C)',color='black', linewidth = 2)
plt.title('Appenrent Temperature & Humidity variation with time(in years)')
plt.legend(loc= 3,fontsize = 16)
plt.show()
```
